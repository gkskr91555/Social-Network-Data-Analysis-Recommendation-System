```python
with open('datainsta.txt', encoding='utf-8') as f:
    instadata = f.read()
```

```python
instadata
```

```python
chunks = instadata.split("\n\n")
chunks = [c for c in chunks if len(c)>3]
```

```python
chunks[4]
```

```python
def parse_chunk(chunk):
    chunk = chunk.strip()
    sep_chunk = chunk.split('\n')
    
    username = sep_chunk[0]
    name = sep_chunk[1]
    no_of_posts = int(sep_chunk[2].split(" post")[0].replace(",",""))
    
    no_of_followers = float(sep_chunk[3].split(" followers")[0].replace(",","").replace("K","").replace("M",""))
    if("K" in sep_chunk[3]):
        no_of_followers = int(no_of_followers * 1000)
    elif("M" in sep_chunk[3]):
        no_of_followers = int(no_of_followers * 1000000)
    else:
        no_of_followers = int(no_of_followers)

    no_of_following = int(sep_chunk[4].split(" following")[0].replace(",","").replace("K","").replace("M",""))
    if("K" in sep_chunk[4]):
        no_of_following = int(no_of_following * 1000)
    elif("M" in sep_chunk[4]):
        no_of_following = int(no_of_following * 1000000)
    else:
        no_of_following = int(no_of_following)

    if(len(sep_chunk)>5):
        categ = sep_chunk[5]
        bio = "\n".join(sep_chunk[6:])
    else:
        categ = "Unknown"
        bio = ""
    
    #print(username,name,no_of_posts,no_of_followers,no_of_following,bio, sep = "\n")
    return {"username": username, "name": name, "no_of_posts": no_of_posts, "no_of_followers": no_of_followers, 
            "no_of_following": no_of_following, "categ" : categ, "bio": bio}
    
#parse_chunk(chunks[0])

    
```

```python
all_chunks = []
for chunk in chunks:
    parsed_chunk = parse_chunk(chunk)
    all_chunks.append(parsed_chunk)
print(all_chunks)
```

```python
import json
s = json.dumps(all_chunks, indent = 4)
with open("instadata.json","w") as f:
    f.write(s)
print(s)
```
