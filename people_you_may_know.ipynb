```python
import json

def load_data(filename):
    with open(filename, "r") as f:
        return json.load(f)

def people_you_may_know(user_id,data):
    user_friends = {}
    for user in data['user']:
        user_friends[user['id']] = set(user['friends'])

    if user_id not in user_friends:
        return[]

    direct_friends = user_friends[user_id]
    suggesitions = {}
    for friend in direct_friends:
        for mutual in user_friends[friend]:
            if mutual != user_id and mutual not in direct_friends:

                #count mutual friends
                suggesitions[mutual] = suggesitions.get(mutual,0) + 1
    sorted_suggesitions = sorted(suggesitions.items(), key=lambda x : x[1], reverse=True)
    return [user_id for user_id, mutual_count in sorted_suggesitions]
    

#load the data
data = load_data("data.json")
user_id = 1
recc = people_you_may_know(user_id, data)
print(recc)
```
