## create json file

```python
obj = { 
    "user":[
        {"id": 1, "name": "Gautam", "friends": [2,3], "liked_pages": [101]},
        {"id": 2, "name": "Sharma", "friends": [1,4], "liked_pages": [102]},
        {"id": 3, "name": "Vishal", "friends": [1], "liked_pages": [101,103]},
        {"id": 4, "name": "Himanshu", "friends": [2], "liked_pages": [104]}
    ],
    "pages": [
        {"id": 101, "name": "Python developers"},
        {"id": 102, "name": "Data Science"},
        {"id": 103, "name": "AI & ML Community"},
        {"id": 104, "name": "Web Dev Hub"}
    ]
}
s = json.dumps(obj)
print(s)
```

```python
file.close()
```

## data visualize

```python
def load_data(filename):
    with open(filename,"r")as f:
        data = json.load(f)
    return data
```

```python
data = load_data("data.json")
```

```python
# write a function to display users and their connection
def display_users(data):
    print("users and their connections:\n")
    for user in data['user']:
        print(f"ID:{user['id']} - {user['name']} is friends with: {user['friends']} and like pages are {user['liked_pages']}")
    print("\nPages information:\n")
    for page in data['pages']:
        print(f"{page['id']}: {page['name']}")

display_users(data)
```
